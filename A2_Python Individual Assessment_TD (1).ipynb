```python
In recent times, bike-sharing has become an increasingly favored and sustainable form of urban mobility. With growing urban congestion and a shift toward eco-conscious living, bicycles serve as an efficient, cost-effective, and environmentally friendly solution for both daily commuters and recreational users. This project analyzes a detailed dataset of bike rentals to uncover the main drivers of user behavior—such as time of use, weather patterns, seasonal trends, and holidays. Identifying these usage patterns will enable the company to refine its operations, better anticipate demand, and develop targeted marketing efforts. Gaining insights into rental behaviors will ultimately support smarter service delivery and improve the overall customer experience.
Gebhart, K., & Noland, R. B. (2014). The impact of weather conditions on bikeshare trips in Washington, DC. Transportation, 41(6), 1205–1225. https://doi.org/10.1007/s11116-014-9540-7
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# specifying a file name
file = 'bikes_data.xlsx'
df = pd.read_excel(file)

# Simulated user input 
user_month = "June"
user_season = "Summer"

# Filter the dataframe using the input
filtered_df = df[(df["Month"] == user_month) & (df["Season"] == user_season)]

# Feedback to user
print(f"\nYou selected {user_month} during the {user_season} season.")
print(f"Total number of entries found: {len(filtered_df)}")
print("Here's a quick look at the data:")
print(filtered_df[["Month", "Season", "Temperature", "Humidity", "RENTALS"]].head())

# Use sns.lmplot, which creates its own figure 
g = sns.lmplot(data=filtered_df, 
               x="Temperature", 
               y="RENTALS", 
               hue="Day of Week", 
               height=6, 
               aspect=1.0, 
               legend=True)  # Let lmplot handle the legend

# Formatting axes and title
g.set(title='Bike Rentals vs Temperature',
      xlabel='Temperature (°C)',
      ylabel='Number of Rentals')

plt.tight_layout
plt.show()        

# Summary Stats
avg_rentals = filtered_df["RENTALS"].mean()
max_temp = filtered_df["Temperature"].max()
min_temp = filtered_df["Temperature"].min()

print(f"\nSummary for {user_month} ({user_season}):")
print(f"- Average rentals: {avg_rentals:.2f}")
print(f"- Temperature range: {min_temp}°C to {max_temp}°C")
```

```python
# Create a list of unique seasons and initialize a dictionary to store average rentals
seasons = df['Season'].unique().tolist()
average_rentals_by_season = {}

# Use a for loop to calculate average rentals for each season
for season in seasons:
    season_df = df[df['Season'] == season]  # Filter data for the season
    avg_rentals = season_df['RENTALS'].mean()
    average_rentals_by_season[season] = round(avg_rentals, 2)

# Print the dictionary with dynamic messaging
print("Average Bike Rentals by Season:\n")
for season, avg in average_rentals_by_season.items():
    print(f"- {season}: {avg} rentals on average")

# Create a list of months and summarize peak rental month
months = df['Month'].unique().tolist()
rentals_by_month = {}

for month in months:
    month_total = df[df['Month'] == month]['RENTALS'].sum()
    rentals_by_month[month] = month_total

# Sort and print the month with the highest total rentals
peak_month = max(rentals_by_month, key=rentals_by_month.get)
print(f"\nPeak rental month: {peak_month} with {rentals_by_month[peak_month]} total rentals.")
```
